```python
import sys
import os
import matplotlib.pyplot as plt
import numpy as np

sys.path.append(os.path.abspath(os.path.join('..')))

from src.sir_model import AgentBasedSIR

os.makedirs('../plots', exist_ok=True)
```

```python
Days = 100
N = 5000
model = AgentBasedSIR(N=N, beta=0.2, gamma=1.0/7.0, C=8)

S, I, R = [], [], []
for _ in range(Days):
    s, i, r = model._get_counts()
    counts, _ = model.step()
    S.append(counts[0])
    I.append(counts[1])
    R.append(counts[2])
```

```python
plt.figure(figsize=(10, 6))
plt.plot(S, label='Susceptible', linewidth=2)
plt.plot(I, label='Infectious', linewidth=2)
plt.plot(R, label='Recovered', linewidth=2)
plt.title('SIR Dynamics (Agent-Based Simulation)')
plt.xlabel('Time (Days)')
plt.ylabel('Population')
plt.legend()
plt.grid(True, alpha=0.3)

plt.savefig('../plots/01_sir_dynamics.png', dpi=300)
plt.show()
print("Plot saved to plots/01_sir_dynamics.png")
```
